```python
import pandas as pd # import libraries
import numpy as np
```

```python
df = pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M') #scrap the data from the wikipedia webpage
df
```

```python
df2 = df[0] #change the extracted data into dataframe
df2
```

```python
df3 = df2[df2['Borough'] != 'Not assigned'].reset_index(drop = True) #drop the rows where column 'Borough' has value of 'Not assigned'
df3
```

```python
len(df3['Postal Code'].unique()) #Check the total number of unique postal codes
```

```python
df4 = df3.groupby(['Postal Code', 'Borough'], as_index = False).agg(lambda x : ", ".join(x)) #Group neighbourhoods in the same borough
df4.head()
```

```python
for index, row in df4.iterrows():                 #For Neighbourhoods == "Not assigned" change them to same as Borough
    if row["Neighbourhood"] == "Not assigned":
        row["Neighbourhood"] = row["Borough"]
        
df4.head()
```

```python
# Check whether it is same as required by the question
column_names = ['Postal Code', 'Borough', 'Neighbourhood']
df5 = pd.DataFrame(columns = column_names)
test_list = ["M5G", "M2H", "M4B", "M1J", "M4G", "M4M", "M1R", "M9V", "M9L", "M5V", "M1B", "M5A"]
for post_code in test_list:
    df5 = df5.append(df4[df4['Postal Code'] == post_code], ignore_index = True)
df5
```
